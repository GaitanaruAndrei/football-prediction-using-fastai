```
# أضف بكل مباراة عدد الأهداف المسجلة كل فريق بالمباريات السابقة
def get_accumlated_goals_scored_by_team(df):
  teams = {key:[0] for key in df.HomeTeam.unique()}
  
  for i, row in df.iterrows():
    
    # add accumlated home team goals based on previous matches
    teams[row.HomeTeam].append(teams[row.HomeTeam][-1] + row.FTHG)
      
    # add accumlated away team goals based on previous matches
    teams[row.AwayTeam].append(teams[row.AwayTeam][-1] + row.FTAG)
  
  
  df['HTGS'] = 0
  df['ATGS'] = 0
  for i, row in df.iterrows():
    
    df.at[i, 'HTGS'] = teams[row.HomeTeam].pop(0)
    df.at[i, 'ATGS'] = teams[row.AwayTeam].pop(0)
  
  return df
```

```
# أضف بكل مباراة عدد نقاط كل فريق خلال المباريات السابقة

def get_match_point(is_home_team, FTR):
  if FTR == 'H':
    return 3 if is_home_team else 0
  elif FTR == 'A':
    return 0 if is_home_team else 3
  else:
    return 1

def get_accumlated_points_by_team(df):
  teams = {key:[0] for key in df.HomeTeam.unique()}
  
  for i, row in df.iterrows():
    
    # add accumlated home team goals based on previous matches
    teams[row.HomeTeam].append(teams[row.HomeTeam][-1] + get_match_point(is_home_team=True, FTR=row.FTR))
      
    # add accumlated away team goals based on previous matches
    teams[row.AwayTeam].append(teams[row.AwayTeam][-1] + get_match_point(is_home_team=False, FTR=row.FTR))
    
  df['HTP'] = 0
  df['ATP'] = 0
  for i, row in df.iterrows():
    
    df.at[i, 'HTP'] = teams[row.HomeTeam].pop(0)
    df.at[i, 'ATP'] = teams[row.AwayTeam].pop(0)
  
  return df
```

```
df_season_19 = get_accumlated_points_by_team(df_season_19)

df_season_18 = get_accumlated_points_by_team(df_season_18)
df_season_17 = get_accumlated_points_by_team(df_season_17)
df_season_16 = get_accumlated_points_by_team(df_season_16)
df_season_15 = get_accumlated_points_by_team(df_season_15)

df_season_14 = get_accumlated_points_by_team(df_season_14)
df_season_13 = get_accumlated_points_by_team(df_season_13)
df_season_12 = get_accumlated_points_by_team(df_season_12)
df_season_11 = get_accumlated_points_by_team(df_season_11)
```

```
# أضف لكل مباراة نتيجة آخر 5 مباريات لكل فريق

from copy import deepcopy

def get_match_result(is_home_team, FTR):
  if FTR == 'H':
    return 'W' if is_home_team else 'L'
  elif FTR == 'A':
    return 'L' if is_home_team else 'W'
  else:
    return 'D'

def get_teams_last_5_games_form(df):
  teams = {key:['M'] for key in df.HomeTeam.unique()} # 'M' since it is the begining of the season
  
  for i, row in df.iterrows():
    
    # add accumlated home team goals based on previous matches
    teams[row.HomeTeam].append(get_match_result(is_home_team=True, FTR=row.FTR))
      
    # add accumlated away team goals based on previous matches
    teams[row.AwayTeam].append(get_match_result(is_home_team=False, FTR=row.FTR))
    
  
  teams2 = deepcopy(teams)
  teams3 = deepcopy(teams)
  teams4 = deepcopy(teams)
  teams5 = deepcopy(teams)   
  for key in teams:
    teams2[key].insert(0,'M')
    
    teams3[key].insert(0,'M')
    teams3[key].insert(0,'M')
    
    teams4[key].insert(0,'M')
    teams4[key].insert(0,'M')
    teams4[key].insert(0,'M')
    
    teams5[key].insert(0,'M')
    teams5[key].insert(0,'M')
    teams5[key].insert(0,'M')
    teams5[key].insert(0,'M')

  # initialize last 5 games to 'M'
  for i in range(1, 6):
    df['HM' + str(i)] = 'M'
    df['AM' + str(i)] = 'M'
  
  for i, row in df.iterrows():
    
    df.at[i, 'HM5'] = teams5[row.HomeTeam].pop(0)
    df.at[i, 'HM4'] = teams4[row.HomeTeam].pop(0)
    df.at[i, 'HM3'] = teams3[row.HomeTeam].pop(0)
    df.at[i, 'HM2'] = teams2[row.HomeTeam].pop(0)
    df.at[i, 'HM1'] = teams[row.HomeTeam].pop(0)
    
    df.at[i, 'AM5'] = teams5[row.AwayTeam].pop(0)
    df.at[i, 'AM4'] = teams4[row.AwayTeam].pop(0)
    df.at[i, 'AM3'] = teams3[row.AwayTeam].pop(0)
    df.at[i, 'AM2'] = teams2[row.AwayTeam].pop(0)
    df.at[i, 'AM1'] = teams[row.AwayTeam].pop(0)
  
  return df
```

```
df_season_19 = get_teams_last_5_games_form(df_season_19)

df_season_18 = get_teams_last_5_games_form(df_season_18)
df_season_17 = get_teams_last_5_games_form(df_season_17)
df_season_16 = get_teams_last_5_games_form(df_season_16)
df_season_15 = get_teams_last_5_games_form(df_season_15)

df_season_14 = get_teams_last_5_games_form(df_season_14)
df_season_13 = get_teams_last_5_games_form(df_season_13)
df_season_12 = get_teams_last_5_games_form(df_season_12)
df_season_11 = get_teams_last_5_games_form(df_season_11)
```

```
# أضف لكل مباراة عدد نقاط الفريق آخر 5 مباريات

df_all_seasons['HTFormPtsStr'] = df_all_seasons['HM1'] + df_all_seasons['HM2'] + df_all_seasons['HM3'] + df_all_seasons['HM4'] + df_all_seasons['HM5']
df_all_seasons['ATFormPtsStr'] = df_all_seasons['AM1'] + df_all_seasons['AM2'] + df_all_seasons['AM3'] + df_all_seasons['AM4'] + df_all_seasons['AM5']

def result_to_points(result):
  if result == 'W':
    return 3
  elif result == 'L':
    return 0
  else:
    return 1

def sum_form_points(form):
  sum = 0
  for f in form:
    sum += result_to_points(f)
  return sum
  
def get_form_points(df):
  
  df['HTFormPts'] = None
  df['ATFormPts'] = None
  for i, row in df.iterrows():
    df.at[i, 'HTFormPts'] = sum_form_points(row.HTFormPtsStr)
    df.at[i, 'ATFormPts'] = sum_form_points(row.ATFormPtsStr)
  
  return df

df_all_seasons = get_form_points(df_all_seasons)
```

```
# أضف لكل فريق كم مرة فاز أو خسر 5 و3 مرات متتالية
def get_win_lose_streaks(df):
  
  df['HTWinStreak3'] = 0
  df['HTWinStreak5'] = 0
  df['HTLossStreak3'] = 0
  df['HTLossStreak5'] = 0

  df['ATWinStreak3'] = 0
  df['ATWinStreak5'] = 0
  df['ATLossStreak3'] = 0
  df['ATLossStreak5'] = 0
  
  for i, row in df.iterrows():
    
    if 'WWW' in row.HTFormPtsStr:
      df.at[i, 'HTWinStreak3'] = 1
    if 'WWWWW' in row.HTFormPtsStr:
      df.at[i, 'HTWinStreak5'] = 1
    
    if 'WWW' in row.ATFormPtsStr:
      df.at[i, 'ATWinStreak3'] = 1
    if 'WWWWW' in row.ATFormPtsStr:
      df.at[i, 'ATWinStreak5'] = 1
      
     
    if 'LLL' in row.HTFormPtsStr:
      df.at[i, 'HTLossStreak3'] = 1
    if 'LLLLL' in row.HTFormPtsStr:
      df.at[i, 'HTLossStreak5'] = 1
      
    if 'LLL' in row.ATFormPtsStr:
      df.at[i, 'ATLossStreak3'] = 1
    if 'LLLLL' in row.ATFormPtsStr:
      df.at[i, 'ATLossStreak5'] = 1
  
  return df

df_all_seasons = get_win_lose_streaks(df_all_seasons)
```

```
df_pred['HTFormPtsStr'] = df_pred['HM1'] + df_pred['HM2'] + df_pred['HM3'] + df_pred['HM4'] + df_pred['HM5']
df_pred['ATFormPtsStr'] = df_pred['AM1'] + df_pred['AM2'] + df_pred['AM3'] + df_pred['AM4'] + df_pred['AM5']

def result_to_points(result):
  if result == 'W':
    return 3
  elif result == 'L':
    return 0
  else:
    return 1

def sum_form_points(form):
  sum = 0
  for f in form:
    sum += result_to_points(f)
  return sum
  
def get_form_points(df):
  
  df['HTFormPts'] = None
  df['ATFormPts'] = None
  for i, row in df.iterrows():
    df.at[i, 'HTFormPts'] = sum_form_points(row.HTFormPtsStr)
    df.at[i, 'ATFormPts'] = sum_form_points(row.ATFormPtsStr)
  
  return df

df_pred = get_form_points(df_pred)
```
